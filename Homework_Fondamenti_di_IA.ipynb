The attributes related with the physical condition are: Calories consumption monitoring (**SCC**), Physical activity frequency (**FAF**), Time using technology devices (**TUE**), Transportation used (**MTRANS**)
variables obtained :
**Gender**, **Age**, **Height** and **Weight**.

**NObesity** values are:

* Underweight
* Normal
* Overweight
* Obesity I

```python
y_train=dt['Weight'].copy()
x_train=dt.drop('Weight',axis=1).copy()

x_test=dtt.drop('Weight',axis=1).copy()
y_test=dtt['Weight'].copy()
xt=x_test.to_numpy()
yt=y_test.to_numpy()
x=x_train.to_numpy()
y=y_train.to_numpy()
```

```python
# implemento la regressione lineare

def linearreg(x, y):
  c= x.T @ x
  c=np.linalg.inv(c)
  c= c @ x.T
  c= c @ y
  return(c)

def predizione(c, x):
  y= x @ c
  return (y)


```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
# TESTING DEL MODELLO

par=linearreg(x,y)
y_learned=predizione(par, xt)

mse = mean_squared_error(yt, y_learned)
#print('mean squared error for linear regression: ', mse)
rmse=np.sqrt(mse)
print('root mean squared error for linear regression: ', rmse)
mae= mean_absolute_error(yt, y_learned)
print('mean absolute error for linear regression: ', mae)
#print(y_learned)
```

```python
#DEFINIZIONE DELLA STRUTTURA DELL'ALBERO DI DECISIONE
class branch:
  def __init__(self,subtree, l):
    self.label=l
    self.subtree=subtree


class decisiontree:
  def __init__(self,a):
    self.a=a
    self.branches=[]
  def addfiglio(self, branch):
    self.branches.append(branch)
```

```python
#IMPLEMENTAZIONE ALBERO DI DECISIONE
import math

def decision_tree_learn(examples, attributes, t, limit, parent_examples):  #returns a tree
# CASO BASE non ci sono piu esempi da analizzare
  if examples.empty:
    return plurality_value(parent_examples, t)


#CASO BASE se hanno tutti la stessa classificazione restituisco il valore dalla classificazioen di uno di essi
  if examples[t].nunique() == 1:
    r= v=examples[t].iloc[0]
    return decisiontree(r)

#CASO BASE sono stati esaminati tutti gli attributi o abbiamo superato l'altezza massima dell'albero
  if not attributes or (limit<0):
    return plurality_value(examples,t)

#CASO RICORSIONE
  else:
  #cerco attributo con maggiore importanza
    i=importance(attributes, examples, t)
    if i=='':
      return plurality_value(examples,t)
    partition=examples.groupby(i)
    tree= decisiontree(i)
    for value,group in partition:
      a=attributes- {i}
      subtree= decision_tree_learn(group,a, t,limit-1, examples)
      tree.addfiglio(branch(subtree, value))
    return tree

#identifica il valore piu comune dell'attributo target
def plurality_value(e, target):
  partition= e.groupby(target)
  i=0
  for value, group in partition:
    righe=group.shape[0]

    if righe> i:
      i=righe
      mcvalue=value
  return decisiontree(mcvalue)

# identifica l'attributo nell insieme attr con il massimo guadagno nel dataset e
def importance (attr, e, target):
  i=0
  a=''
  for column in attr:
    g=guadagno (e, column, target)
    if g>i:
      i=g
      a=column
  return a

#calcolo del guadagno
def guadagno(e, a , target):
    res= entropy(e,target)-remainder(e,a,target)
    return res

#calcolo dell'entropia
def entropy(e, target):
  if (e[target].nunique() == 1):
    return 0
  sum=e[target].sum()
  p=sum/(e.shape[0])
  return -p*math.log2(p)-(1-p)*math.log2(1-p)


#calcolo del remainder
def remainder(df,column,target):
  r=0
  partition=df.groupby(column)
  for value,group in partition:
    r= r+(group.shape[0]/df.shape[0])*entropy(group,target)
  return r
```
